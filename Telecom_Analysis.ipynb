## Handling Missing Values

The dataset initially contained missing values in four key columns: **Offer**, **Internet Type**, **Churn Category**, and **Churn Reason**. Addressing these missing values is crucial for accurate analysis and model building.


***the following columns have missing values***:

* **Offer:** 3877 missing values
* **Internet Type:** 1526 missing values
* **Churn Category:** 5174 missing values
* **Churn Reason:** 5174 missing values

```python
data['Offer'] = data['Offer'].fillna('No Offer')
data['Internet Type'] = data['Internet Type'].fillna('No Internet Service')

print("Missing values after imputing 'Offer' and 'Internet Type':")
display(data[['Offer', 'Internet Type']].isna().sum())
```

### Actions Taken

- **Offer Column**  
  Missing values (3,877 entries) were imputed with the string **"No Offer"**.  
  This approach assumes that a blank *Offer* entry signifies the absence of any promotional offer for that customer.

- **Internet Type Column**  
  Missing values (1,526 entries) were imputed with the string **"No Internet Service"**.  
  This suggests that customers without an *Internet Type* do not subscribe to internet services.


```python
data['Offer']
```

```python
data['Churn Category'] = data['Churn Category'].fillna('No Churn')
data['Churn Reason'] = data['Churn Reason'].fillna('No Churn')

print("Missing values after imputing 'Churn Category' and 'Churn Reason':")
display(data[['Churn Category', 'Churn Reason']].isna().sum())
```

**Actions Taken:**

*   **Churn Category:** All 5174 missing values were imputed with the string **'No Churn'**. This is based on the understanding that these missing values correspond to customers who have not churned, hence they do not have a churn category.
*   **Churn Reason:** Similarly, all 5174 missing values were imputed with the string **'No Churn'**. This indicates that non-churning customers do not have a specific reason for churn.



```python
data[['Churn Category', 'Churn Reason']]
```

**Current State:**
Following these imputation steps, all previously identified columns (`Offer`, `Internet Type`, `Churn Category`, and `Churn Reason`) now have **0 missing values**. The dataset is now prepared for further exploratory data analysis and modeling, with a clear understanding of the meaning behind the previously missing entries.
